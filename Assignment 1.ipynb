# 100 best movies ever made according to  Sight & Sound

Sight & Sound is a monthly UK film magazine published by British Film Institute (BFI) since 1932 ( Link 1).  Every decade Sight & Sound organizes an international poll of critics to decide what films are the greatest ever made.  According to Nick James in order to choose the respondents to the poll they , “approached more than 1,000 critics, programmers, academics, distributors, writers and other cinephiles, and received (in time for the deadline) precisely 846 top-ten lists that between them mention a total of 2,045 different films” (Link 2).  For the last 50 years Citizen Kane (1941) by Orson Welles was the leader of the chart. But in 2020 Vertigo (1958) by Alfred Hitchcock took the first place. 

In order to answer the question about what movies are represented in that rating the dataset with the following columns from Link 1 was made: title, director, year, country, length, colour, votes and about. The last column "sex" was added manually. that column represents the sex of the movie director. Here is the exploration of the dataset followed by insights and conclusions!

## What directors has the most of the pictures presented in the ratings?

```python
#See statistics of distribution depending on the Directors
s=films.Director
counts=s.value_counts()
dir_films=pd.DataFrame( { 'counts': counts })
dir_films=dir_films.reset_index()
dir_films=dir_films.rename(columns={"index": "Director "})
dir_films.head(10)
```

## Conclusion: Godard, Bergman and Hitchcock.
Godard, Bergman and Hitchcock have four movies each represented in top 100 rating. Bresson, Dreyer, Welles, Coppola and Tarkovsky are taking the second place with three movies each.

## What coutries are movies coming from?

```python
#See statistics of distribution depending on the Country
s=films.Country
counts=s.value_counts()
percent=s.value_counts(normalize=True)
percent100=s.value_counts(normalize=True).mul(100).round(1).astype(str)+'%'
country_films=pd.DataFrame( { 'counts': counts, "per": percent, "per100": percent100 })
country_films=country_films.reset_index()
county_films=country_films.rename(columns={"index": "Country "})
country_films.head(10)
```

## Conclusion: USA is a leader with 32%.

32% of the movies in the rating are coming from USA, 19% from France. Third place is divided between former USSR and Japan. These countries contributed six movies each tot eh rating. 

## What genre is the most presented in the rating?

```python
#See statistics of distribution depending on the Genre
s=films.Genre
counts=s.value_counts()
percent=s.value_counts(normalize=True)
percent100=s.value_counts(normalize=True).mul(100).round(1).astype(str)+'%'
genre_films=pd.DataFrame( { 'counts': counts, "per": percent, "per100": percent100 })
genre_films=genre_films.reset_index()
genre_films=genre_films.rename(columns={"index": "Genre "})
genre_films.head(10)
```

## Conclusion: Drama.
Dramatic movies are the majority in the rating.

## How many women directors are in the rating?

```python
#See statistics of distribution depending on the Sex
s=films.Sex
counts=s.value_counts()
percent=s.value_counts(normalize=True)
percent100=s.value_counts(normalize=True).mul(100).round(1).astype(str)+'%'
sex_films=pd.DataFrame( { 'counts': counts, "per": percent, "per100": percent100 })
sex_films=sex_films.reset_index()
sex_films=sex_films.rename(columns={"index": "Sex "})
sex_films
```

```python
# Pie Chart of data distribution of the sex Feminine vs Masculine

labels = 'Men', 'Women '
sizes = sex_films['counts']
colors = ["#64b5cd", "#da8bc3"]
explode = (0.6, 0.1)  # explode 1st slice

# Plot
plt.pie(sizes, explode=explode, labels=labels, colors=colors,
autopct='%1.1f%%', shadow=True, startangle=140)
plt.title('Men vs Women Directors', fontsize=16)
plt.axis('equal')
plt.show()
```
